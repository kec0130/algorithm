### 561. Array Partition I

```python
# 2172 ms
class Solution:
    def arrayPairSum(self, nums: List[int]) -> int:
        nums.sort()
        sum = 0
        while len(nums) > 0:
            sum += min(nums[0], nums[1])
            nums = nums[2:]
        return sum
```

```python
# 284 ms
class Solution:
    def arrayPairSum(self, nums: List[int]) -> int:
        nums.sort()
        num_list = [nums[i] for i in range(len(nums)) if i % 2 == 0]
        return sum(num_list)
```
